__1) Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.__

$$ k = 85; n = 100; p = 0.8; q = 0.2 $$

$$ P(k=85)=C^{k}_{n}\cdot p^{k} \cdot q^{n-k} = C^{85}_{100}\cdot p^{85} \cdot q^{100-85} = \frac{100!}{85!\cdot(100-85)!} \cdot 0.8^{85} \cdot 0.2^{15} = 0.048 $$

__Вероятность того, что стрелок попадет в цель ровно 85 раз равна:__ $\,\,\,0.048$

$$P_{5000}(2) = \frac{\lambda^{m}}{m!}\cdot e^{-\lambda}=\frac{2^{2}}{2!}\cdot e^{-2} = 2 \cdot e^{-2} = 0.27067$$

__3) Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?__

$$ k = 70; n = 144; p = 0.5; q = 0.5 $$

$$ P(k=70)=C^{k}_{n}\cdot p^{k} \cdot q^{n-k} = C^{70}_{144}\cdot p^{70} \cdot q^{144-70} = \frac{144!}{70!\cdot(144-70)!} \cdot 0.5^{70} \cdot 0.5^{74} = 0.0628 $$

__Вероятность, что орел выпадет ровно 70 раз из 144 подбрасываний равна:__ $\,\,\,0.0628$

__Вероятность того, что хотя бы один извлечённый мяч белый:__

- для одного белого мяча из первого ящика количество благоприятных исходов равно:

$$ C^{1}_{3} = \frac{3!}{1!\cdot(3-1)!} = \frac{3}{1} = 3  $$

$$ C_{1} = 7 \cdot 3 = 21 $$

- для двух белых мячей из первого ящика количество благоприятных исходов равно:

$$ C^{2}_{7} = \frac{7!}{2!\cdot(7-2)!} = \frac{6\cdot7}{2} = 21 $$

$$ C^{0}_{3} = \frac{3!}{0!\cdot(3-0)!} = 1  $$

$$ C_{2} = 21 $$

- для одного белого мяча из второго ящика количество благоприятных исходов равно:

$$ C^{1}_{9} = \frac{9!}{1!\cdot(9-1)!} = \frac{9}{1} = 9 $$

$$ C^{1}_{2} = \frac{2!}{1!\cdot(2-1)!} = \frac{11}{1} = 2  $$

$$ C_{3} = 9 \cdot 2 = 18 $$

- для двух белых мячей из второго ящика количество благоприятных исходов равно:

$$ C^{2}_{9} = \frac{9!}{2!\cdot(9-2)!} = \frac{8\cdot9}{2} = 36 $$
